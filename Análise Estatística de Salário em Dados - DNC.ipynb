<span style="font-size:30px"><b>Etapa 02</b></span>

```python
contagem_cargos = df['job_title'].value_counts()

plt.Figure(figsize=(10, 5))
sns.barplot(x=contagem_cargos.head(10).values, y=contagem_cargos.head(10).index)
plt.title('Top 10 Cargos mais Comuns', fontsize=20)
plt.xlabel('Número de Profissionais', fontsize=12)
plt.ylabel('Cargo', fontsize=12)

plt.show()
```

```python
contagem_experiencia = df['experience_level'].value_counts()
print('Distribuição por Nível de Experiência:')
print(contagem_experiencia)

plt.Figure(figsize=(10, 5))
sns.countplot(data=df, x='experience_level', order=['EN', 'MI', 'SE', 'EX'])

plt.title('Distribuição por Nível de Experiência', fontsize=20)
plt.xlabel('Nível de Experiência', fontsize=12)
plt.ylabel('Número de Profissionais', fontsize=12)
plt.xticks([0, 1, 2, 3], ['Junior (EN)', 'Pleno (MI)', 'Sênior (SE)', 'Executivo (EX)'], rotation=0)

plt.show()
```

```python
contagem_tamanho = df['company_size'].value_counts()
print('Distribuição por Tamanho da Empresa:')
print(contagem_tamanho)

plt.Figure(figsize=(10, 5))
sns.countplot(data=df, x='company_size', order=['S', 'M', 'L'])

plt.title('Distribuição por Tamanho da Empresa', fontsize=20)
plt.xlabel('Tamanho da Empresa', fontsize=12)
plt.ylabel('Número de Profissionais', fontsize=12)
plt.xticks([0, 1, 2], ['Pequena (S)', 'Média (M)', 'Grande (L)'], rotation=0)

plt.show()
```

<span style="font-size:30px"><b>Etapa 03</b></span>

```python
resumo_salarial = df['salary_in_usd'].describe()

print('Resumo Estatístico dos Salários (USD):')
print(resumo_salarial)
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['salary_in_usd'], bins=30, kde=True)

plt.axvline(x=df['salary_in_usd'].mean(), color='red', linestyle='--', label=f'Média: ${df['salary_in_usd'].mean():,.2f}')
plt.axvline(x=df['salary_in_usd'].median(), color='cyan', linestyle='--', label=f'Mediana: ${df['salary_in_usd'].median():,.2f}')

plt.title('Distribuição dos Salários (USD)', fontsize=20)
plt.xlabel('Salário (USD)', fontsize=12)
plt.ylabel('Número de Profissionais', fontsize=12)
plt.legend()

plt.show()
```

```python
plt.Figure(figsize=(10, 5))
sns.boxplot(data=df,
            x='experience_level',
            y='salary_in_usd',
            order=['EN', 'MI', 'SE', 'EX'],
            whiskerprops=dict(color='blue', linewidth=2),
            capprops=dict(color='blue', linewidth=2),
            medianprops=dict(color='cyan', linewidth=2))

plt.title('Distribuição Salarial por Nível de Experiência', fontsize=20)
plt.xlabel('Nível de Experiência', fontsize=12)
plt.ylabel('Salário (USD)', fontsize=12)
plt.xticks([0, 1, 2, 3], ['Junior (EN)', 'Pleno (MI)', 'Sênior (SE)', 'Executivo (EX)'], rotation=0)

plt.show()
```

<span style="font-size:30px"><b>Etapa 04</b></span>

```python
media_salarial_por_pais = df.groupby('country')['salary_in_usd'].mean()
top10_paises = media_salarial_por_pais.sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=top10_paises.values, y=top10_paises.index)

plt.suptitle('COM OUTLIERS', fontsize=10)
plt.title('Top 10 Países com Maiores Médias Salariais (USD)', fontsize=20)
plt.xlabel('Média Salarial (USD)', fontsize=12)
plt.ylabel('País', fontsize=12)

plt.show()
```

```python
paises_grafico_ant = top10_paises.index
df_filt = df[df['country'].isin(paises_grafico_ant)]
count_nos_top10 = df_filt['country'].value_counts()

print('Número de Profissionais por país (no top 10 por Média):')
print(count_nos_top10)
```

```python
contagem_paises = df['country'].value_counts()
paises_com_amostra_suficiente = contagem_paises[contagem_paises >= 10].index
df_filtrado_robusto = df[df['country'].isin(paises_com_amostra_suficiente)]
mean_final = df_filtrado_robusto.groupby('country')['salary_in_usd'].mean().sort_values(ascending=False).head(10)

plt.Figure(figsize=(10, 5))
sns.barplot(x=mean_final.values, y=mean_final.index)

plt.suptitle('SEM OUTLIERS', fontsize=10)
plt.title('Top 10 Países por Média Salarial', fontsize=20)
plt.xlabel('Média Salarial (USD)', fontsize=12)
plt.ylabel('País', fontsize=12)

plt.show()

```

<span style="font-size:30px"><b>Etapa 05</b></span>

```python
colunas_para_correlacao = df[['salary_in_usd', 'work_year', 'years_of_experience']]
matriz_correlacao = colunas_para_correlacao.corr()

plt.figure(figsize=(10, 5))
sns.heatmap(matriz_correlacao, annot=True, cmap='coolwarm', fmt='.2f')

plt.title('Matriz de Correlação', fontsize=20)

plt.show()
```
